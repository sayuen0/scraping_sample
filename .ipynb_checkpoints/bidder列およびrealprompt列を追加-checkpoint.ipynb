```python
import pandas as pd
df = pd.read_csv("一ページ目商品詳細.csv",index_col=0)
df
```

```python
# 乱数モジュール
import random 
# biddersの代わりに1から40くらいまでの整数の擬似乱数を作る(もっかい取得するのめんどくさいので)
random_list =[]

# 100件の乱数リスト。擬似データなので本物のbiddersに置き換えてください
for i in range(0,100):
    random_list.append(random.randint(1,40))
random_list #中身。ごく稀に1があるみたい
    
```

```python
# bidders列を登録
df["bidders"] = random_list
```

```python

#df["bidders"] == 1 
# DataFrameの列名に対して比較演算をすると、TrueかFalseかの100件のSeriesができる
# そこに1をかけると、Trueを1、Falseを0と皆して掛け算してくれる
is_single_bidder =( df["bidders"] == 1)*1
is_single_bidder
```

```python
# realprompt列は
# is_prompt列が0なら0が入るし、
# is_single_bidder列がFalse(入札者が一人でないなら)0が入るし、
# どっちも1のときのみ1が入る
df["realprompt"] = df["is_prompt"] * is_single_bidder
print(df)

df.to_csv("bidder列およびrealprompt列追加済み_一ページ目商品詳細.csv")
```
