## 1. 検索ワードから表示件数ごとのHTMLを保存


### 関数の説明

sleep_(): 休憩を入れる(捕まらないため)

メイン。
query_name: ここに入れた名前でフォルダ名がセーブされる → airpods
url : ここに検索するURLを直接入力。ただし &b=数値は削除しておくこと。また&=100としておくべき。

### 手順

urlに検索したい商品の1ページ目を代入

## 2. DataFrameを59個作成

### 手順
ここでは関数を定義しただけ。特別ここで操作することは無し

### 関数の説明

add_name(dataframe, soup):dataframeとbeautifulsoupを代入して、商品名列が追加されたbeautifulsoupを

```python
#2019-03-12みたいな日付けを作る
def formatted_today():
    import datetime
    now = datetime.datetime.now()
    today = "{0:%Y-%m-%d}".format(now)
    return today

def print_line():
    print("---------------------------------------------------------------------------------------------------------------------------------------")

def sleep_(sec):
    print("過剰な連続リクエスト回避のsleep:"+str(sec)+"秒")
    sleep(sec)
    

def get_product_count_list():
#     [1,101,201,301,...]
# そのページで何番目の商品番号から始まるかのリスト。
#  &b=101　のようにURLに含める
    page_list = []
    for i in range(0, 12):
        page_list.append(i*100+1)
    return page_list



def get_save(query_name, url):

    print("クローリングを開始します")
    print("アーティスト名: "+query_name)
    


#   [1,101,201,301,...]
    page_list =get_product_count_list()
    
#   iにはpage_listのインデックス、すなわち[0,1,2,3,4,5,...]
#  page　には値、すなわち[1,101,201,301]が入る
    for i, page in enumerate(page_list):
        
        print(query_name+"の"+str(i)+" ページ目のクローリングを開始...")

        r = requests.get(url+"&b="+str(page))
        soup = BeautifulSoup(r.text, "html.parser")
        print(str(i)+"ページ目取得完了...")
# 日付けからディレクトリ作成
        date = formatted_today()
        combined_path = query_name+"/"+date
        os.makedirs(combined_path, exist_ok=True)
#     書き込み
        with open (combined_path+"/"+"ticket-data"+str(i)+".html", mode="w", encoding="utf-8") as fw:
            fw.write(soup.prettify())
            print(query_name+"の"+str(i)+"ページ目の保存中...")

        print(query_name+"の"+str(i)+" ページ目のクローリングを完了しました。")
        sleep_(2)


# get_save(url, name)

url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=airpods+第2世代+新品&va=airpods+第2世代+新品&n=100"
name ="ヤフオク"
get_save(name,url)


print_line()
print("全アーティストのクローリングを完了しました")
print_line()
```

```python
# df = pd.DataFrame()
# url ="./airpods/2019-09-11/ticket-data1.html"
# soup = BeautifulSoup(open(url), "html.parser")
# 公演名: name


def add_name(dataframe,soup):
#     names = list(map(lambda x: convert_zenkaku_to_hankaku(x.string.strip().replace("\n","").replace("\u3000","")), soup.select(".Product__titleLink") ))
    names = soup.select(".Product__titleLink")
#     namesは20件のaタグの「リスト」が入っている。リストにはstringできない
# 　　 namesの各要素にstringしてやる必要がある
    list = []
#     文字の加工
    for name in names:
        list.append(name.string.strip().replace("\n","").replace("\u3000",""))
    dataframe["商品名"] = list
    return dataframe

# add_name(df)

# df
    
#     ラムダ式：配列の各要素に対して同じ操作をしたものを返す
#     各要素に対して soup.stringを使う→タグのなかのテキストだけ取れる


# セレクタ
#  .クラス
# 　#id
```

```python

def add_price_value(dataframe,soup):
    prices = soup.select(".Product__priceValue")
    list = []
    for price in prices:
        list.append(int(price.string.strip().replace("\n","").replace("\u3000","").replace("円","").replace(",","")))
    dataframe["価格"] = list
    return dataframe

    
# add_price_value(df)
# df
```

```python
# CSV/日付/ticket-data1.csv, ..ticket-data2.csv
#  concat→concatenated_csv.csv


# DataFrameを12回作る

for i in range(0, 12):
    page_list= get_product_count_list()
    

df_list = []

for i,name in  enumerate(page_list):
    df = pd.DataFrame()
    url ="./ヤフオク/2019-09-11/ticket-data"+str(i)+".html"
    soup = BeautifulSoup(open(url), "html.parser")
    df  = add_name(df,soup)
    df = add_price_value(df,soup)
    df_list.append(df)

# [DataFrame, DataFrame, DataFrame, DataFrame, ...] 59個
d = pd.concat(df_list,ignore_index=True)
d.to_csv("CSV/concated_csv.csv",)
```
